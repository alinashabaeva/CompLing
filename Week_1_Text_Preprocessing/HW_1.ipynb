Похоже, Mystem с дизамбигуацией работает несколько лучше, однако различия незначительны. Рассмотрим несколько наиболее интересных примеров. Во-первых, "стоять" и "стоить". MyStem с дизамбигуацией правильно относит "стоит" к н.ф."стоить" (если проверить по контексту). В то время как Mystem без дизамбигуации относит этот глагол к  форме "стоять". Во-вторых, "разумеется" mystem без дизамбигуации относит к глаголу "разуметься", а с дизамбигуацией - к вводн.слову "разумеется" (правильный вариант). С другой стороны, есть примеры, в которых mystem без дизамбигуации работает лучше. Например, "теплее" mystem без дизамбигуации относит к прил. "теплый" (правильно по контексту), а mystem с дизамбигуацией - к наречию "тепло". Итак, одназначно сказать, что лучше, сложно, но, возможно, Mystem с дизамбигуацией работает немного лучше (особенно если анализируется большой объем данных).
